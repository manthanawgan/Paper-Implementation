```python
import matplotlib.pyplot as plt
import numpy as np
```

### Relu

```python
def ReLu(x):
  return np.maximum(0, x)

x = np.linspace(-10, 10)
y = ReLu(x)

print(y)

plt.title('Rectified linear unit')
plt.plot(x, y)
```

### Sigmoid

```python
def Sigmoid(x):
  y_01 = 1/(1 + np.exp(-x))
  return y_01

x = np.linspace(-10, 10)
y = Sigmoid(x)

print(y)

plt.title('Sigmoid')
plt.plot(x, y)
```

### Hyperbolic tangent


```python
def Tanh(x):
  y_02 = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
  return y_02

x = np.linspace(-15,15)
y = Tanh(x)

plt.title('Tanh')
plt.plot(x, y)
```

### Leaky ReLu

```python
def Leaky_ReLu(x):
  return np.maximum(0.01 * x, x)

x = np.linspace(-1000, 100)
y = Leaky_ReLu(x)

plt.title('Leaky ReLU')
plt.plot(x, y, label='Leaky ReLU')
plt.xlabel('x')
plt.ylabel('Leaky_ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

### Softmax

```python
def Softmax(x):
  exp_x = np.exp(x - np.max(x))
  return np.divide(exp_x, np.sum(exp_x))

x = np.linspace(-10, 10)
y = Softmax(x)

plt.title('Softmax')
plt.plot(x, y)
```
